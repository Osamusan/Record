```
import re
import zipfile
import urllib.request
import os.path,glob
 
#青空文庫から、ダウンロードしたいURLを入力する

URL = 'https://www.aozora.gr.jp/cards/000148/files/2669_ruby_6341.zip' # 岡本一平著並画『探訪画趣』 序 
def main():
    download_text = download(URL)
    text = convert(download_text)
    print(text)
    
    with open('test.txt', 'w', encoding = 'utf-8') as F:  # 出力をファイル'test.txt'に収める
       F.write(text)
 
def convert(download_text):
    binarydata = open(download_text, 'rb').read()
    text = binarydata.decode('shift_jis')
    
 
    # ルビ、注釈などの除去
    text = re.split(r'\-{5,}', text)[2]
    text = re.split(r'底本：', text)[0]
    text = re.sub(r'《.+?》', '', text)
    text = re.sub(r'［＃.+?］', '', text)
    text = text.strip()
    text = re.sub(r'。','。\r\n',text)
    return text
 
def download(url):
 # データファイルをダウンロードする
 zip_file = re.split(r'/', url)[-1]
 
 if not os.path.exists(zip_file):
     print('Download URL')
     print('URL:',url)
     urllib.request.urlretrieve(url, zip_file)
 else:
     print('Download File exists')
 
 # フォルダの生成
 dir, ext = os.path.splitext(zip_file)
 if not os.path.exists(dir):
     os.makedirs(dir)
 
 # zipファイルの展開
 zip_obj = zipfile.ZipFile(zip_file, 'r')
 zip_obj.extractall(dir)
 zip_obj.close()
 
 # zipファイルの削除
 os.remove(zip_file)
 
 # テキストファイルの抽出
 path = os.path.join(dir,'*.txt')
 list = glob.glob(path)
 return list[0]
 
if __name__ == "__main__":
    main()
```

```
# 形態素解析結果のcsvファイルの作成

tagger = MeCab.Tagger(path)
datafile = open('test.txt', encoding = 'utf-8')

outfile = open('kekka.csv', 'w')

for line in datafile:
  line = line.rstrip()
  line = line.lstrip('　')  # 文頭の全角スペース削除
  line_mecab = tagger.parse(line)
 # print(line_mecab)
  words = line_mecab.split('\n')
  for i in range(0,len(words)-1):
    lex = words[i].split('\t')
#    print(lex)
    for j in range(0,len(lex)):
        print(lex[j] + ',', end ='')
        outfile.write(lex[j] + ',')
    outfile.writelines('\n')   
          
    print('\r\n')
    
#files.download('kekka.csv')
```

```
# Concordanceの作成

allcontext =[]
entry = []
context_width = 10  # 前後の文脈長の決定

out_F = open('kwic.csv', 'w')

with open('kekka.csv') as F:
  for line in F:   
    line = line.rstrip()
    line = line.lstrip() 
    line_list = line.split(",")
    
    if line_list[0] == 'EOS': # ' EOS'の削除
       continue
 
    lex = line_list[0]
    yomi = line_list[1]
    goiso = line_list[3]
    pos1 = line_list[4]
          
    entry = [lex, yomi, goiso, pos1]
    allcontext.append(entry)

context_size = len(allcontext)
for item in range(0,context_size): #context_size kaisu

  #左側文脈を作成  
  left_context = ''
  for j in range(item-context_width, item):
    if j < 0:
      continue
    words = allcontext[j][0]
    context_word = words
    left_context += context_word

  #右側文脈を作成
  right_context = ''
  for j in range(item+1, item+1+context_width):
    if j >= context_size:
      continue
    words = allcontext[j][0]
    context_word = words
    right_context += context_word

  words = allcontext[item][0]
  key = words
  lexime = allcontext[item][2]
  pron = allcontext[item][1]
  pos = allcontext[item][3]

  output = ','.join([lexime, pron, left_context, key, right_context, pos])
  print(output)
  
  out_F.write(output+'\n')

#files.download('kwic.csv')
```
